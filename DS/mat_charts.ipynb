```python
# visualizations 
#  matplotlib, seaborn, plotly-express 
# !pip install matplotlib

import matplotlib.pyplot as plt
```

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
```

```python
# bar plot - 
```

```python
plt.bar(x= fruits, height= sales)
```

```python
plt.bar(x= fruits, height= sales,width=0.4)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.bar(x= fruits, height=sales,width=0.4)
plt.show()
```

```python
plt.figure(figsize=(4,3))
# plt.bar(x= fruits, height=sales,width=0.4,color='red')
plt.bar(x= fruits, height=sales,width=0.4,color=['red','yellow','blue','green'])
plt.title("Sales of Fruits")
plt.xlabel("Fruits")
plt.ylabel("Sales")
plt.show()
```

```python
# pie chart 
plt.pie(x=sales, labels=fruits,autopct='%1.2f%%', colors=['yellow','pink','green','skyblue'])
plt.show()
```

```python
# line chart
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperature = [22, 24, 23, 26, 25]

plt.plot(days, temperature, marker='o')
plt.title('Weekly Temperature')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# scatter plot 
import numpy as np
x = np.array([12, 45, 7, 32, 89, 54, 23, 67, 14, 91])
y = np.array([99, 31, 72, 56, 19, 88, 43, 61, 35, 77])
z=np.array([99, 31, 72, 60, 10, 80, 40, 60, 15, 70])
```

```python
plt.scatter(x=x,y=y, marker="o",c='purple',s=z)
plt.show()
```
