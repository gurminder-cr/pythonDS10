```python
import seaborn as sns  # pip install seaborn
```

```python
import matplotlib.pyplot as plt 
```

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
tips.day.value_counts()
```

```python
plt.scatter(x=tips.loc[:,'total_bill'], y=tips.loc[:,'tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
a= sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', size='size')
a.set_title('Total Bill vs Tips')
a.set_xlabel("Total Bill")
a.set_ylabel("Tips")
```
