```python
import pandas as pd 
import seaborn as sns 

df= sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('deck',axis=1, inplace=True)
```

```python
df.head()
```

```python
df[df.age.isnull()]
```

```python
df.who.value_counts()
```

```python
df_man= df[df.loc[:,'who']=='man']
```

```python
df_man
```

```python
man_age_mean=df_man.age.mean()
man_age_mean
```

```python
df_man[df_man.age.isnull()]
```

```python
df_man_index= df_man[df_man.age.isnull()].index
df_man_index
```

```python
df.loc[df_man_index,'age']=man_age_mean
```

```python
df.isnull().sum()
```

```python
df.loc[df_man_index,:]
```
